```python
import numpy as np

class LinearRegression:
  bias_ = 0
  weigth_ = 0
  def fit(self, x, y):
    x_np = np.array(x)
    y_np = np.array(y)
    sum_x = np.sum(x_np)
    sum_y = np.sum(y_np)
    sum_x2 = np.sum(x_np**2)
    sum_xy = np.sum(x_np*y_np)
    n = len(x_np)
    self.weigth_ = n*sum_xy - sum_x*sum_y
    self.weigth_ /= n*sum_x2 - sum_x**2
    self.bias_ = (sum_y - self.weigth_*sum_x)/n
  def predict(self, value):
    return self.weigth_*value + self.bias_
```

```python
x =  [80,150,30,50,120,60,110,110]
y = [ 16,30,12,10,24,18,20,25]

linear_regression = LinearRegression()
linear_regression.fit(x,y)
print(linear_regression.bias_)
print(linear_regression.weigth_)
linear_regression.predict(70)
```

```python
import matplotlib.pyplot as plt
x_line = np.arange(1,150)
y_line = linear_regression.predict(x_line)
plt.scatter(x,y,c="green")
plt.plot(x_line,y_line,c="red")
plt.show()
```
